```python
from keras.models import Sequential
from keras.layers import Dense
```

The goal of this exercise is to combine layers and create a neural network.

1. Create a neural network for regression with the following architecture and return `print(model.summary())`:
   - Five input variables.
   - Hidden layer 1: 8 neurons and sigmoid as the activation function.
   - Hidden layer 2: 4 neurons and sigmoid as the activation function.
   - Output layer: 1 neuron. Find the adapted activation function.

```python
model = Sequential([
    Dense(units=8, activation='sigmoid', input_shape=(5,)),
    Dense(units=4, activation='sigmoid'),
    Dense(units=1, activation='linear')
])
print(model.summary())
```
